```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/simple-house-price-prediction/house_price.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
corr = data.corr()
sns.heatmap(corr, cmap = 'Wistia', annot= True);
```

```python
sns.lmplot(x='area',y='price',data=data,aspect=2,height=6)
plt.xlabel('Area')
plt.ylabel('Price')
```
